```python
script_list = ["HDFCBANK.NS","SBIN.NS","INFY.NS","ICICIBANK.NS","AXISBANK.NS","MARUTI.NS","WIPRO.NS","BHARTIARTL.NS","ASIANPAINT.NS","DIVISLAB.NS","HDFC.NS","HCLTECH.NS","TECHM.NS","KOTAKBANK.NS","TITAN.NS","TATACONSUM.NS","LT.NS","UPL.NS","ITC.NS","EICHERMOT.NS","TCS.NS","ADANIPORTS.NS","CIPLA.NS","DRREDDY.NS","RELIANCE.NS","ULTRACEMCO.NS","M&M.NS","BAJAJFINSV.NS","GRASIM.NS","NESTLEIND.NS","INDUSINDBK.NS","HEROMOTOCO.NS","POWERGRID.NS","NTPC.NS","BRITANNIA.NS","SUNPHARMA.NS","BAJAJ-AUTO.NS","BAJFINANCE.NS","HINDALCO.NS","BPCL.NS","SHREECEM.NS","TATASTEEL.NS","JSWSTEEL.NS","COALINDIA.NS","HINDUNILVR.NS","TATAMOTORS.NS","IOC.NS","ONGC.NS"]

a = []
df = []
for i in script_list:
    a.append(yf.Ticker(i))
for j in range(len(script_list)):
    df.append(a[j].history(start = "2011-01-01", end = "2022-06-01"))

```

```python
for i in range(len(script_list)):
    Macd = ta.trend.MACD(close = df[i]["Close"])
    df[i]["MACD"] = Macd.macd()
    df[i]["Signal"] = Macd.macd_signal()
    Stoch = ta.momentum.StochRSIIndicator(close = df[i]["Close"])
    df[i]["%k"] =  Stoch.stochrsi_k()
    df[i]["%d"] = Stoch.stochrsi_d()
    MA = ta.trend.SMAIndicator(close = df[i]["Close"], window = 200)
    df[i]["SMA"] = MA.sma_indicator()
```
